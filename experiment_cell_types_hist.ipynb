# Importing and checking tsv

```python
# Load the TSV file into a DataFrame
file_path = '/home/marina/Downloads/experiment_report_2023_12_19_11h_23m.tsv'

# Assuming the file is tab-separated, use read_csv with delimiter='\t'
df = pd.read_csv(file_path, delimiter='\t')

```

## Check if 'Target of assay' and 'Target gene symbol' are equal 

```python
# Check if the values in "Target of assay" column match the values in "Target gene symbol" column
target_assay_values = df['Target of assay']
target_gene_values = df['Target gene symbol']

# Check if the values in both columns are the same
are_equal = target_assay_values.equals(target_gene_values)

if are_equal:
    print("Values in 'Target of assay' column are the same as in 'Target gene symbol' column.")
else:
    print("Values in 'Target of assay' column are different from 'Target gene symbol' column.")
```

```python
# Filter rows where 'Target of assay' values are different from 'Target gene symbol' values
different_values = df[df['Target of assay'] != df['Target gene symbol']]
    
if not different_values.empty:
        print("Rows with different values in 'Target of assay' and 'Target gene symbol':")
        print(different_values.head())  # Display the first few rows with different values
else:
        print("No rows found with different values in 'Target of assay' and 'Target gene symbol'.")
```

# Grouping the data by 'Target of assay' and counting unique values in 'Accessions' for each group

```python
# Grouping the data by 'Target of assay' and counting unique values in 'Accessions' for each group
accessions = df.groupby('Target of assay')['Accession'].nunique()
```

```python
more_than=accessions[accessions>3]
print(f'We have {len(more_than)} Tfs with more than 3 accesions')
more_than=accessions[accessions>5]
print(f'We have {len(more_than)} Tfs with more than 5 accesions')
more_than=accessions[accessions>7]
print(f'We have {len(more_than)} Tfs with more than 7 accesions')
```

# Create Histograms for Accesions numbers

```python
# Convert the result to a DataFrame for easier plotting
accessions_df = accessions.reset_index(name='Unique Accessions Count')


# Create a histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=accessions_df, x='Unique Accessions Count', kde=True)
plt.title('Histogram of Unique Accessions for Each Target of Assay')
plt.xlabel('Number of Unique Accessions')
plt.ylabel('Frequency')
plt.show()
```

```python
# Filter the DataFrame to include only values up to 10
accessions_df_filtered = accessions_df[accessions_df['Unique Accessions Count'] <= 10]

# Create a histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=accessions_df_filtered, x='Unique Accessions Count', bins=10, kde=True)
plt.title('Histogram of Unique Accessions Counts (Up to 10) for Each Target of Assay')
plt.xlabel('Number of Unique Accesions')
plt.ylabel('Frequency')

# Set x-axis limit to show values up to 10
plt.xlim(0, 10)

plt.show()
```

```python
# Create CDF plot using Seaborn
sns.ecdfplot(accessions, stat='count', complementary=True)

# Set x-axis limit to 0-100
plt.xlim(0, 20)

# Set plot labels and title
plt.xlabel('Accessions')
plt.ylabel('Number of TFs')
plt.title('Cumulative Distribution Function (CDF)')

# Set number of ticks along the x-axis and y-axis
plt.locator_params(axis='x', nbins=30)  # Adjust the number of ticks on the x-axis (in this case, 10 ticks)
plt.locator_params(axis='y', nbins=30)  # Adjust the number of ticks on the y-axis (in this case, 10 ticks)

# Show plot
plt.show()
```
