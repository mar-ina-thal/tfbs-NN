```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_metric_values(file_path, additional_title=""):
    # Read the CSV file into a DataFrame
    try:
        df = pd.read_csv(file_path)
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file_path}")
        return
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return

    # Calculate TP/FP ratio for each threshold
    try:
        df['tp_fp_ratio'] = df['TP'] / df['FP']
    except KeyError:
        print(f"Error: 'TP' or 'FP' columns not found in file {file_path}")
        return

    # Find the threshold with the maximum TP/FP ratio
    max_ratio_row = df.loc[df['tp_fp_ratio'].idxmax()]

    # Plot 'TP', 'TN', 'FP', 'FN' for the best threshold with different colors
    colors = ['blue', 'green', 'red', 'purple']
    plt.bar(['TP', 'TN', 'FP', 'FN'], max_ratio_row[['TP', 'TN', 'FP', 'FN']], color=colors)
    
    # Set the title with the additional string
    plt.title(f'{additional_title} : Metric Values for Best Threshold {max_ratio_row["Threshold"]}')
    plt.show()

    # Print the best threshold and corresponding ratio
    print(f"Best Threshold: {max_ratio_row['Threshold']}, Max TP/FP Ratio: {max_ratio_row['tp_fp_ratio']}")


```

```python
#Mapping between original strings and numerical values:
#{'CTCF Binding Site': 0, 'Enhancer': 1, 'Open chromatin': 2, 'Promoter': 3, 'TF binding': 4, 'not_regulatory': 5}
file_path = '/home/marina/CNN3_results_200_batch_64_kernel_5_chrom_split_true/0_results.csv'
additional_title = "CTCF Binding Site"
plot_metric_values(file_path, additional_title)
```
